```python
import pyfpgrowth
import pickle
with open('allup.pkl','rb') as f:
    allup=pickle.load(f)
with open('alldown.pkl','rb') as f:
    alldown=pickle.load(f)

print(len(allup))
print(len(alldown))
```

```python
up_patterns = pyfpgrowth.find_frequent_patterns(allup,13)
up_patterns
```

```python
down_patterns = pyfpgrowth.find_frequent_patterns(alldown, 13)
down_patterns
```

```python
up_rules = pyfpgrowth.generate_association_rules(up_patterns, 0.7)
up_rules
```

```python
down_rules = pyfpgrowth.generate_association_rules(down_patterns, 0.7)
down_rules
```
