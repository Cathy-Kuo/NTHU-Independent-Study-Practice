```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import gym
from collections import namedtuple
import numpy as np
import matplotlib
import random
import math
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
is_ipython = 'inline' in matplotlib.get_backend()
if is_ipython:
    from IPython import display
```

```python
replay = namedtuple('replay',('state','action','reward','next_state'))

class Replay_buffer():
    
    def __init__(self , buffer_size):
        self.buffer_size = buffer_size
        self.buffer = np.zeros(  [buffer_size] , dtype = replay)
        self.index = 0
        self.cur_size = 0
        
    def push(self,experience):
        self.buffer[self.index] = experience
        self.index = (self.index+1)%self.buffer_size
        if self.cur_size < self.buffer_size:
            self.cur_size += 1
    def sample(self,batch_size):
        sample_index = np.random.choice(np.arange(self.cur_size),size=batch_size,replace=False)
        return self.buffer[sample_index]

    
class DQN(nn.Module):
    
    def __init__(self , NUM_OBS , NUM_ACTION):
        super(DQN,self).__init__()
        self.L1 = nn.Linear(NUM_OBS,64)
        self.L2 = nn.Linear(64,64)
        #self.L3 = nn.Linear(120,120)
        self.out = nn.Linear(64,NUM_ACTION)
    
    def forward(self,x):
        x = F.relu(self.L1(x))
        x = F.relu(self.L2(x))
        #x = F.relu(self.L3(x))
        x = self.out(x)
        return x
```

```python
NUM_OBS = 8
NUM_ACTION = 4


num_epsoide = 2000
network = DQN(NUM_OBS , NUM_ACTION)
target_network = DQN(NUM_OBS , NUM_ACTION)
target_network.load_state_dict(network.state_dict())
loss_fn = torch.nn.MSELoss(reduction='mean')
buffer = Replay_buffer(10000)
EPS_START = 1.0
EPS_END = 0.05
#EPS_DECAY = 2000
EPS_DECAY = 0.9975
steps_done = 0
batch_size = 128
use_cuda = torch.cuda.is_available()
optimizer = torch.optim.Adam(network.parameters(),lr=0.0001)
eps_threshold = EPS_START

env = gym.make('LunarLander-v2').unwrapped
reward_history = []
shaped_reward_history = []
loss_history = []
try:
    for e in range(num_epsoide):
        state = env.reset()
        state = torch.tensor(state).float()
        DONE = False
        acc_reward = 0
        shaped_acc_reward = 0
        time_step = 0
        for t in range(1000):
            if e % 10 == 0:
                env.render(mode='rgb_array') 
            
            if random.random() >= eps_threshold:
                action = torch.argmax(network(state))
            else:
                action = torch.tensor(env.action_space.sample(),dtype=torch.long)
                
            next_state , reward , done , _ = env.step(action.item())
            time_step += 1
            steps_done += 1
            acc_reward += reward
            reward = float(reward) - abs(next_state[1])*2 - abs(next_state[0])
            
            shaped_acc_reward += reward
            next_state = torch.tensor(next_state).float()
            if done:
                DONE = True
                next_state = None
            
            experience = replay(state , torch.tensor([action]).long(), \
                                torch.tensor([reward]) , next_state)

            buffer.push(experience)
            state = next_state
        
            if buffer.cur_size >= batch_size:

                batch = buffer.sample(batch_size=batch_size)
                batch = replay(*zip(*batch))

                non_final_mask = torch.tensor(tuple(map(lambda s: s is not None,
                                              batch.next_state)), dtype=torch.uint8)
                
                
                
                if use_cuda:
                    non_final_next_state = torch.cat([s for s in batch.next_state if s is not None]).view(-1,NUM_OBS).cuda()
                    state_batch = torch.cat(batch.state).view(-1,NUM_OBS).cuda()
                    reward_batch = torch.cat(batch.reward).cuda()
                    action_batch = torch.cat(batch.action).view(-1,1).cuda()
                    state_action_value = network(state_batch).gather(1,action_batch).cuda()
                    next_state_values = torch.zeros(batch_size).cuda()
                    next_state_values[non_final_mask] = target_network(non_final_next_state).max(1)[0].cuda()

                    expected_state_values = next_state_values + reward_batch.view(-1)
                    #expected_state_values = expected_state_values
                    state_action_value = network(state_batch).gather(1,action_batch).view(-1).cuda()
                    
                    

                else:
                    non_final_next_state = torch.cat([s for s in batch.next_state if s is not None]).view(-1,NUM_OBS)
                    state_batch = torch.cat(batch.state).view(-1,NUM_OBS)
                    reward_batch = torch.cat(batch.reward)
                    action_batch = torch.cat(batch.action).view(-1,1)
                    state_action_value = network(state_batch).gather(1,action_batch)
                    next_state_values = torch.zeros(batch_size)
                    next_state_values[non_final_mask] = target_network(non_final_next_state).max(1)[0]
                    expected_state_values = next_state_values + reward_batch.view(-1)
                    #expected_state_values = expected_state_values
                    state_action_value = network(state_batch).gather(1,action_batch).view(-1)
                    
                    
                

                loss = loss_fn(state_action_value,expected_state_values)

                optimizer.zero_grad()
                loss.backward()
                optimizer.step()
                loss_history.append(loss.item())
            if e %5 == 0:
                target_network.load_state_dict(network.state_dict())
            if done == True:
                break
        eps_threshold = max(eps_threshold * EPS_DECAY , EPS_END)
        reward_history.append(acc_reward)
        shaped_reward_history.append(shaped_acc_reward)
        if e % 50 == 0 and e > 0:
            print(sum(reward_history[-50:])/50 , eps_threshold)
        if(sum(reward_history[-100:])/100 >= 200):
            print("done in eps {}".format(e))
            break
finally:
    env.close()
```
